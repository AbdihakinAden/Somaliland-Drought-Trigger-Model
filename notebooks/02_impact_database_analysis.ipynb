```python
import sys, os
sys.path.append(os.path.abspath(".."))

import pandas as pd

# ------------------------------
# Load SPI means and impact data
# ------------------------------
spi = pd.read_csv("../data/processed/somaliland_spi_2020_2023.csv")
impact = pd.read_csv("../data/processed/validation_impact_2020_2023.csv")

# ------------------------------
# Merge SPI mean with observed impacts
# ------------------------------
df = spi.merge(
    impact[["region", "year", "season", "OBS"]],
    on=["region", "year", "season"],
    how="left"
)

# ------------------------------
# Clean display â€” ONLY these columns:
# region, year, season, SPI mean, observed drought (0/1)
# ------------------------------

df_clean = df[["region", "year", "season", "mean", "OBS"]].sort_values(
    ["region", "year", "season"]
)

df_clean.head()
```

```python
df_clean.to_csv("../data/processed/impact_spi_clean.csv", index=False)
```
