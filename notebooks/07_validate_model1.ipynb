```python
# 07_validate_model1.ipynb
import sys, os
sys.path.append(os.path.abspath(".."))

import pandas as pd
from pathlib import Path

from src.model1_weighted import compute_model1
from src.validation import validate_thresholds

# config
MERGED_CSV = "../data/processed/merged_for_models.csv"
OUT_METRICS = "../results/validation/model1_metrics.csv"
THRESHOLDS = [25, 30, 40]
OBS_COL = "OBS"

# load merged data
df = pd.read_csv(MERGED_CSV)
print("Rows loaded:", len(df))

# compute Model 1
df1 = compute_model1(df, out_col="DHI1")

# validate using DHI1 column
metrics_df = validate_thresholds(df1, value_col="DHI1", thresholds=THRESHOLDS, obs_col=OBS_COL)

# save predictions (optional) and metrics
pred_out = "../results/validation/model1_predictions.csv"
Path(pred_out).parent.mkdir(parents=True, exist_ok=True)
df1.to_csv(pred_out, index=False)

Path(OUT_METRICS).parent.mkdir(parents=True, exist_ok=True)
metrics_df.to_csv(OUT_METRICS, index=False)

print("Model 1 validation metrics:")
display(metrics_df)
print("Predictions saved to:", pred_out)
print("Metrics saved to:", OUT_METRICS)
```
